## What to think about when making figures


1. Who is your target audience?
2. What is it that you want to convey (is a figure the best way to do it)?
3. Choose the best type (lines, points, bars, pie charts etc.)?
4. Choose the best range and scales for the data you are presenting (e.g. linear, logarithmic etc.).
5. Avoid trying to convey too much information at once.
6. Use colors and patterns effectively (and consider color blindness).
7. Make sure axis, labels, legends and annotations are readable (and avoid unnecessary text).
8. Try to adopt a consistent style throughout your report/scientific presentation/paper.
9. Use a (concise) caption in reports or papers.
10. ...

## Basics

This diagram shows some of the different naming conventions that are used to describe a Matplotlib plot.

- **Title:** The title of a Matplotlib plot is the heading that appears at the top of the figure.
- **Axes:** It is the region where data is plotted on a Matplotlib plot.
- **Axis:** These are the number-line-like objects. They take care of setting the graph limits
  and generating the ticks (the marks on the axis) and tick labels (strings labeling the ticks).
- **Grid:** It is the dotted line that divides the axes into multiple equal square spaces.
- **Label:** It is the notation for axis. For instance, we can define the x-axis label
  and y-axis label names.
- **Legend:** It is the notation for plots in the figure.
  It is often represented with distinct colors, shapes, and names.
- **Figure:** A figure in Matplotlib means the whole window in the user interface.
  Within this figure, there can be subplots that help to arrange plots in a regular grid
  where we need to specify the number of rows and columns and the number of the plot.

The most straightforward plot is a simple line plot that is done using ```plot()```. Let's look at a simple example.

````python
# Dummy Data
X = [1, 2, 3, 4]
Y = [2, 4, 6, 8]

plt.plot(X, Y)        # plt.plot() is used for plotting a line plot
plt.show()            # a normal python script will also need plt.show() for displaying the plot
````

Now we add several improvements, including a title, and labels. We will change the color and linewidth and also experiment with changing the size of the figure. Then we add an annotation using an arrow. Finally we add another line and a legend.

````python
# Dummy Data
X = np.array([1, 2, 3, 4])
Y = np.array([2, 4, 6, 8])

A = np.array([1, 2, 3, 4])
B = np.array([1, 2, 3, 4])

# mpl.rcParams.update({'font.size': 15})  # This sets the fontsize and labelsize for the entire notebook
# mpl.rc('xtick', labelsize=15) 
# mpl.rc('ytick', labelsize=15) 

# Defining the figure shape
plt.figure(figsize = (10, 4)) # Width is 10 and height is 4

# plot() is used for plotting a line plot
plt.plot(X, Y, color = 'green', linewidth = 5)
plt.plot(A, B, color = 'red', linewidth = 5)

# Adding annotations
plt.annotate('This is a line', xy=(2.5, 5.5), xytext=(1.5, 7),
             arrowprops={'facecolor':'black', 'shrink':0.1}, 
             fontsize=15
             )

# Adding title, xlabel and ylabel
plt.title('A Basic Line Plot', fontsize=15) #Title of the plot
plt.xlabel('X-axis', fontsize=15) #X-Label
plt.ylabel('Y-axis', fontsize=15) #Y-Label
plt.xticks(fontsize=15)
plt.yticks(fontsize=15)

# Giving labels to each line (to be shown in the legend)
plt.legend(['Green Line', 'Red Line'], loc='lower right', fontsize=15) # Note that we could also have added the names as labels to the 'plot' function and only used plt.legend()

# show() is used for displaying the plot, in principle this is not necessary in jupyter notebook, but it is good practice to keep it.
plt.show()
````

### Some more basics: multiple panels

Now we plot the two lines in separate panels.

````python
# Defining the figure shape
fig, ax = plt.subplots(1, 2) # One row and two columns

# plot() is used for plotting a line plot
ax[0].plot(X, Y, color = 'green', linewidth = 2)
ax[1].plot(A, B, color = 'red', linewidth = 2)

# show() is used for displaying the plot
plt.show()
````

But we also want axis labels and a shared axis.

````python
# Defining the figure shape
fig, ax = plt.subplots(1, 2, sharey=True, constrained_layout=True) # One row and two columns sharing the y-axis

# plot() is used for plotting a line plot
ax[0].plot(X, Y, color = 'green', linewidth = 5, label='Green line')
ax[1].plot(A, B, color = 'red', linewidth = 5, label='Red line')
ax[0].legend()
ax[1].legend()

# Adding title, xlabel and ylabel
fig.suptitle('A Basic Line Plot') # Title of the plot (for the entire set of subplots)
fig.supxlabel('X-axis') # X-Label
fig.supylabel('Y-axis') # Y-Label 

# show() is used for displaying the plot
plt.show()
````

````python
# Defining the figure shape
fig, ax = plt.subplots(2, 3, sharey=False, sharex=True, constrained_layout=True,
                       figsize = (10, 4)) # One row and two columns sharing the y-axis

# plot() is used for plotting a line plot
ax[0,0].plot(X, Y, color = 'green', linewidth = 5, label='Green line')
ax[1,0].plot(A, B, color = 'red', linewidth = 5, label='Red line')
ax[0,1].plot(A, A*A, color = 'green', linewidth = 5, linestyle = ':')
ax[1,1].plot(X, Y*Y, color = 'red', linewidth = 5, linestyle = '--')
ax[0,2].plot(A, A-10, color = 'green', linewidth = 5, linestyle = ':')
ax[1,2].plot(X, Y-10, color = 'red', linewidth = 5, linestyle = '--')

# Adding title, xlabel and ylabel
fig.suptitle('A Basic Line Plot') #Title of the plot (for the entire set of subplots)
fig.supxlabel('X-axis') #X-Label
fig.supylabel('Y-axis') #Y-Label 

# show() is used for displaying the plot
plt.show()
````

## Plotting 2D

There are many different ways one can present data. This ranges from the lines/curves seen above, the scatter plots, bar plots, pie charts and histograms. After the line plot example, we here look at the other types.

### A scatter plot

````python
# Libraries/Modules import conventions
import numpy as np
# matplotlib has already been imported previously. 
# The global parameters for the font size and tick sizes are also adopted. 

# Making the random number generator always generate the same random numbers
np.random.seed(10)

# Dummy Data
x = np.arange(50)
y = x + 10 * np.random.randn(50)

# scatter() is used for plotting a scatter plot
plt.scatter(x, y)

# Adding title, xlabel and ylabel
plt.title('A Basic Scatter Plot') # Title of the plot
plt.xlabel('X-axis') # X-Label
plt.ylabel('Y-axis') # Y-Label

# show() is used for displaying the plot
plt.show()
````

The ```scatter()``` function also allows for different sizes and colors for the symbols. Like this random example. 

````python
# Making the random number generator always generate the same random numbers
np.random.seed(10)

# Dummy Data
x = np.arange(50)
y = x + 10 * np.random.randn(50)

# Defining sizes and colors
sizes = np.abs(np.random.randn(50)) * 100
colors = np.random.randint(0, 50, 50)

# scatter() is used for plotting a scatter plot
plt.scatter(x, y, s=sizes, c=colors)

# Adding title, xlabel and ylabel
plt.title('A Colorful Scatter Plot') # Title of the plot
plt.xlabel('X-axis')                 # X-Label
plt.ylabel('Y-axis')                 # Y-Label

# Adding a line at y=0 and at x=0
plt.axhline(0, c='lightgrey', linestyle='dashed')
plt.axvline(0, c='lightgrey', linestyle='dashed')

# show() is used for displaying the plot
plt.show()
````

### A bar plot

````python
# Dummy Data
x = ['Year 1', 'Year 2', 'Year 3', 'Year 4','Year 5']
y1 = [235, 554, 582, 695, 545]
y2 = [100, 200, 500, 600, 800]
width = 0.35  # the width of the bars

# Making the plot for y1 list's data and plotting
p1 = plt.bar(x, y1, width)
# p1 = plt.barh(x, y1, width)

# Stacking the y2 list's data at top and plotting
p2 = plt.bar(x, y2, width, bottom = y1)
# p2 = plt.barh(x, y2, width, left = y1)

# legend() is used for displaying the plot legend
plt.legend(['y1','y2'])

# Adding title, xlabel and ylabel
plt.title('A Stacked Bar Plot') # Title of the plot
plt.xlabel('X-axis') # X-Label
plt.ylabel('Y-axis') # Y-Label

# show() is used for displaying the plot
plt.show()
````

### A pie chart

````python
# Dummy data
labels = ['Year 1', 'Year 2', 'Year 3', 'Year 4','Year 5']
values = [235, 695, 554, 550, 545]

# Defining a pie that is to explode outside of the pie chart
explode = (0, 0.1, 0, 0, 0)  # only "explode" the 2nd slice (i.e. 'Year 2')

# pie() is used for plotting a pie chart and 'explode' property is used to explode a pie
# plt.pie(values, labels=labels)
plt.pie(values, labels=labels, explode=explode, startangle=90, counterclock=False)

plt.show()
````

### And a histogram

````python
# Making the random number generator always generate the same random numbers
np.random.seed(10)

# Preparing random data
x = np.random.normal(size = 1000)

# hist() is used for plotting a histogram
plt.hist(x, density = True, bins = 30)   # Density=True normalises the histogram area to 1

# Adding title, xlabel and ylabel
plt.title('A Histogram Plot') # Title of the plot
plt.xlabel('X-axis') # X-Label
plt.ylabel('Y-axis') # Y-Label

# show() is used for displaying the plot
plt.show()
````

### Contour and quiver plot examples

Here we give some further examples of types of plot, in particular the (filled) contour plot and quiver plot.  

````python
# First define the linearly spaced X and Y arrays
xlist = np.linspace(-3.0, 3.0, 100)
ylist = np.linspace(-3.0, 3.0, 100)

# Then use the numpy meshgrid routine to create a regular grid; define Z on this grid  
X, Y = np.meshgrid(xlist, ylist)
Z = np.sqrt(X**2 + Y**2)

cp = plt.contourf(X, Y, Z)
plt.colorbar(cp) # Add a colorbar to a plot
plt.title('Filled Contours Plot')
plt.xlabel('x (cm)')
plt.ylabel('y (cm)')
plt.show()
````

````python
# define the grid again, now in one line
x,y = np.meshgrid(np.arange(-2, 2, .2), np.arange(-2, 2, .2))
z = x*np.exp(-x**2 - y**2)

# define the gradient of Z, the spacing is .2,.2 in X and Y 
v, u = np.gradient(z, .2, .2)

plt.title('A quiver Plot')
plt.quiver(x,y,u,v)
plt.xlabel('x')
plt.ylabel('y')
plt.show()
````

## 3D plots

There are many ways to create a 3D plot in matplotlib. We show the use of the ```axes()``` method.

````python
# Axes3D is needed for plotting 3D plots
from mpl_toolkits.mplot3d import Axes3D

# For interactive matplotlib sessions, turn on the matplotlib inline mode
%matplotlib inline

mpl.rcParams.update({'font.size': 12})
mpl.rc('xtick', labelsize=12) 
mpl.rc('ytick', labelsize=12) 

# Dummy data
x = [1, 2, 3, 4, 5] # X-coordinates
y = [1, 2, 3, 4, 5] # Y-coordinates
z = [4, 10, 20, 5, 3] # Z-ccordinates

# Defining figure
fig = plt.figure(figsize = (8, 6), dpi = 90)

# Making 3D Plot using plot3D()
ax = plt.axes(projection = '3d')
ax.plot3D(x, y, z)

# Setting Axis labels
ax.set_xlabel('X-Axis')
ax.set_ylabel('Y-Axis')
ax.set_zlabel('Z-Axis')

# Showing the plot
plt.show()
````

````python
x,y = np.meshgrid(np.arange(-2, 2, .2), np.arange(-2, 2, .2))
z = x*np.exp(-x**2 - y**2)

def plotmodel(angle1, angle2):
    fig = plt.figure(figsize = (8, 6), dpi = 90)
    ax = plt.axes(projection = '3d')
    surf = ax.plot_surface(x, y, z, cmap=mpl.cm.coolwarm)
    ax.view_init(angle2,angle1)
# Setting Axis labels
    ax.set_xlabel('X-Axis')
    ax.set_ylabel('Y-Axis')
    ax.set_zlabel('Z-Axis')
    fig.colorbar(surf, shrink=0.5, aspect=5)
    
@interact
def rotate(angle1=(0,180), angle2=(-90,90)):
        plotmodel(angle1, angle2)
        
plt.show()
````
